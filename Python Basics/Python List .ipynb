### Accessing items from a list

```python
# Indexing
l = [1,2,3,4,5]
# Positive Indexing
print(l[0])
# Negative Indexing
print(l[-1])
```

```python
# Indexing in 2d list
l = [1,3,4,5,[3,5]]
#Negative Indexing
print(l[-1][-2])
# Positive Indexing
print(l[4][0])
```

```python
# Indexing in 3d list 
l = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(l[0][1][0])
print(l[0][0][1])
```

```python
# Slicing
l = [1,2,3,4,5,6]
print(l[0:3])
```

```python
l[-1:-6:-2]
```

```python
l[-3:]
```

### Adding Items in the list

```python
# append : To add one item in the list 
l = [1,2,3,4,5]
l.append(6)
print(l)
l = [1,2,3,4,5]
l.append([6,7])
print(l)
```

```python
# extend : To add multiple items in the list
l = [1,2,3,4,5]
l.extend([6,7,8])
print(l)
l = [1,2,3,4,5]
l.extend('delhi')
print(l)
```

```python
# Insert
l = [1,2,3,4,5]
l.insert(5,6)
print(l)
```

### Deleting items from a list

```python
# del
l = [1,2,3,4,5]
print(l)
del(l[-1])
print(l)
del(l)
print(l)
```

```python
#remove --> removes first 5
l = [1,2,3,4,5,6,5,7,5,7,2,3,4,5]
l.remove(5)
print(l)
```

```python
# POP 
l = [1,2,3,4,5,6,7,8,9]
l.pop()
print(l)
l.pop(0)
print(l)
```

```python
# clear
l = [1,2,3,4,5,6,7,8,9]
l.clear()
print(l)
```

### List Functions

```python
# General functions len/min/max/sorted
l = [1,2,3,4,5,6]
print(len(l))
print(min(l))
print(max(l))
print(sorted(l,reverse = True))
```

```python
# count 
l = [1,2,3,1,3,4,5,6,2]
l.count(2)
```

```python
# index
l = [1,2,3,4,1,3,4,2,3,5,6,7,3,2]
l.index(2)
```

```python
# reverse 
l = [1,2,5,7,10]
# when there is no output then the operation is permanent
l.reverse()
print(l)
```

```python
# sort vs sorted
l = [2,1,5,7,0]
print(l)
print(sorted(l))
print(l)
l.sort()
print(l)
```

```python
# copy --> Shallow copy
l = [2,1,5,7,0]
print(l)
print(id(l))
l1 = l.copy()
print(l1)
print(id(l1))
```

### List Comprehension

List Comprehension provides a concise way of creating lists.

newlist = [expression for item in iterable if condition == True]




#### Advantages of List Comprehension

- More time-efficient and space-efficient than loops.
- Require fewer lines of code.
- Transforms iterative statement into a formula.

```python
# Add 1 to 10 numbers to a list
l = []
for i in range(1,11):
    l.append(i)
print(l)
```

```python
# List comprehension
l = [i for i in range(1,11)]
print(l)
```

```python
# scalar multiplication on a vector
v = [2,3,4]
s = -3
# [-6,-9,-12]
print(v*s)
l = [i*s for i in v]
print(l)
```

```python
# Prepare a list of squares from given list
l = [1,2,3,4,5,6,7,8,9]
[i**2 for i in l]
```

```python
# print all numbers divisible by 5 in range of 1 to 50
[i for i in range(1,51) if i%5 == 0]
```

```python
# find languages that starts with letter p
languages = ['java','python','php','c','javascript']
[lang for lang in languages if lang.startswith('p')]
```

```python
# nested if with list comprehension
basket = ['apple','guave','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']
[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```python
# print a (3,3) matrix using list comprehension - > nested list comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# cartesian product 
l1 = [1,2,3,4]
l2 = [5,6,7,8]
[i*j for i in l1 for j in l2]
```

### 2 ways to traverse a list
- itemwise
- indexwise

```python
# itemwise
for i in [1,2,3,4,5]:
    print(i)
```

```python
# indexwise
l = [1,2,3,4,5]
for i in range(len(l)):
    print(l[i])
    
```

### Zip Function 

```python
#The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together.

# If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.
```

```python
# write a program to add items of 2 lists indexwise
l1 = [1,2,3,4]
l2 = [-1,-2,-3,-4]

print(list(zip(l1,l2)))

[i+j for i,j in list(zip(l1,l2))]

# [i+j for i in l1 for j in l2 if i+j ==0]
```
